```python
import matplotlib.pyplot as plt;
import numpy as np;

k_shtr, w_shtr = np.loadtxt('disp0.txt', unpack=True)

fig, (g1) = plt.subplots(1, 1)

plt.grid(True) 
plt.xlabel('$k$')
plt.ylabel('$\omega$') 
             
g1.scatter(k_shtr, w_shtr, 1)
```
